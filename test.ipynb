```python
def arrange_ongdu(cards):
    first_set, second_set = [], []
    for i, card in enumerate(cards):
        if i < len(cards) / 2:
            first_set.append(card)
        else:
            second_set.append(card)
    return first_set, second_set
```

```python
arrange_ongdu(cards)
```
